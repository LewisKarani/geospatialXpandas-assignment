```python
file_path = r'C:\Users\HP\Documents\my data/201902-fordgobike-tripdata.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df.index
```

```python
spec_stations = df.groupby(['start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude']).size().reset_index(name='ride_totals')
spec_stations.column = ['stat_id', 'stat_name', 'lat', 'longt', 'ride_totals']

popular_statn = spec_stations.loc[spec_stations['ride_totals'].idxmax()]
show_popular = popular_statn['start_station_name']
mpop_how_many = popular_statn['ride_totals']
unpopular_statn = spec_stations.loc[spec_stations['ride_totals'].idxmin()]
show_unpopular = unpopular_statn['start_station_name']
upop_how_many = unpopular_statn['ride_totals']
average_rides_per_stat = spec_stations['ride_totals'].mean()
```

```python
#spec_stations
```

```python
print(f'The most popular station was: {show_popular} with {mpop_how_many} rides!')
print(f'The most unpopular station was: {show_unpopular} with {upop_how_many} rides!')
print(f'The average ride per station was: {average_rides_per_stat} rides!')
```

```python
import folium
```

```python
map_exmp = folium.Map(location=[37.77, -122.42], zoom_start=12, tiles='CartoDB Positron')
def color_marker_identifier(ride_totals):
    if ride_totals <1000:
        return 'red'
    elif ride_totals <5000:
        return 'green'
    else:
        return 'orange'
    
for idx, row in spec_stations.iterrows():
    
    marker_color = color_marker_identifier(row['ride_totals'])
    
    marker = folium.Marker(location=[row['start_station_latitude'], row['start_station_longitude']], 
                           icon=folium.Icon(icon='car', color=marker_color), 
                           popup = f"Station: {row['start_station_name']}<br>Rides: {row['ride_totals']}"
                          )
    marker.add_to(map_exmp)

map_exmp
```
